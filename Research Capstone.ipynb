Hypothesis- The Data Science Admissions process should eliminate offering EA calls on the website. Instead the team should drive applications through email facebook and webinars exclusively (If students ask for an EA call we can offer it). We can therefore save AC’s time have them focus on pushing people through the Admissions funnel.

Testing process- To solve for  this we will be using an A/B test that affects splits the funnel into webinar only and calls offered on the website. The point will be to  see if we can afford to only talk to qualified applicants more often, by putting people in a group setting. 

A Group- For half the people on the website, in emails and on facebook the Call To Action will  where it was to schedule a call it will now be to visit a webinar that happens everyday of the week. The purpose of that webinar will be drive qualified applicants to apply 

B Group- For the other half, the Call to Action will be schedule a call with and Ed advisor through all parts of the funnel. 

Evaluation- There will be three elements to evaluate the experiment. I’ll explain more in detail

1) EA Call held rate to Apply rate vs Webinar engaged to Apply rate
2) Applied to Accepted rate of each funnel 
3) Mode or Average time spent in funnel 

We want to see that students who are raising their hands in one on one calls and in Webinars are actually applying to the program.  Comparing these two ratios will give us and idea of how each funnel is performing. If they are close to equal it proves our hypothesis. If calls drive applications more we should stay the same and if the Webinar drives more applications than we definitely need to switch to webinars exclusively. 
We need to be able to test the quality of each funnel by checking the Applied to Accepted rate of each funnel. Specifically because it shows us that the applicant was strong enough to get accepted to the program. We cannot control if they enroll, but we have more control over them getting accepted. A higher rate here either way shows us that we are on the right track.
We need to track how long it takes for someone to go from interested to enrolled or silent. Here the general rule is that the faster the better. Granted here there might be some outliers that skews the mean so we should use multiple measures of central tendency.

These three measures will be a strong gauge of exactly how the experiment is going. We are looking for either no statistical significance among the group or some that favors a no call model if it shows that in all three measures we are good to go for the change.
